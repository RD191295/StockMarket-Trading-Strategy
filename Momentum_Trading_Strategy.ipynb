```python
import yfinance as yf
import pandas as pd
```

```python
stocks = pd.read_html("https://en.wikipedia.org/wiki/NIFTY_50")[1].Symbol
```

```python
stocks.head(5)
```

```python
stocks = stocks + '.NS'
```

```python
stocks = stocks.to_list()
```

```python
df = yf.download(stocks,start='2010-01-01')['Close']
```

```python
df
```

```python
monthly_price = df.resample('ME').last()
```

```python
monthly_price
```

```python
momentum = monthly_price.pct_change(12).shift(1)
```

```python
portfolio = pd.DataFrame(index = monthly_price.index, columns=monthly_price.columns)
```

```python
portfolio
```

```python
for date in momentum.index[12:]:
    pf_date = momentum.loc[date]
    top20 = pf_date.nlargest(20).index
    portfolio.loc[date,top20] = 1/20
```

```python
top20
```

```python
portfolio.loc[date,top20]
```

```python
returns = monthly_price.pct_change().shift(-1)
```

```python
strategy_returns = (portfolio* returns).sum(axis = 1)
```

```python
(1+strategy_returns).cumprod().plot(title = 'cross_section_strategy returns')
```
